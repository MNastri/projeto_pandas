# **Ajustes**

```python
df_temp = df.loc[df['North America'].isnull() | \
                 df['Europe'].isnull() | \
                 df['Japan'].isnull() | \
                 df['Rest of World'].isnull() | \
                 df['Global;'].isnull()]
```

```python
df_temp
```

```python
df_temp = df
```

```python
df_temp.dropna(subset=['North America'], inplace=True)
```

```python
df_temp
```

```python
df_temp['North America'].isnull().sum()
```

```python
df_temp = df_temp.rename(columns={'Global;': 'Global'})
df_temp
```

```python
df_temp['Global'] = df_temp['Global'].str.replace(';', '').astype(float)
df_temp['Global']
```

# **Faturamento por GÃªnero e paÃ­s**

```python
genre_na = df_temp.groupby('Genre')['North America'].sum().sort_values(ascending=False)
genre_europe = df_temp.groupby('Genre')['Europe'].sum().sort_values(ascending=False)
genre_japan = df_temp.groupby('Genre')['Japan'].sum().sort_values(ascending=False)
genre_remaining_world = df_temp.groupby('Genre')['Rest of World'].sum().sort_values(ascending=False)
genre_global = df_temp.groupby('Genre')['Global'].sum().sort_values(ascending=False)
```
